```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import os
```

```python
df = pd.read_csv("diabetes.csv")
```

```python
df
```

```python
columns = [
    "Pregnancies", "Glucose", "BloodPressure", "SkinThickness", "Insulin",
    "BMI", "DiabetesPedigreeFunction", "Age", "Outcome"
]
```

```python
output_folder = "output_charts"
os.makedirs(output_folder, exist_ok=True)
```

```python
for col in df.columns:
    plt.figure(figsize=(6, 4))
    df[col].value_counts().sort_index().plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f"Bar Chart - {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.tight_layout()
    plt.savefig(f"output_charts/bar_chart_{col}.png")
    plt.show()  
    plt.close()

```

```python
for col in df.columns:
    plt.figure(figsize=(6, 4))
    plt.hist(df[col], bins=20, color='orange', edgecolor='black')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
 

df = pd.read_csv('diabetes.csv')
 

for col in df.columns:
    if df[col].nunique() > 10:
        
        binned_col = pd.cut(df[col], bins=10)
        value_counts = binned_col.value_counts().sort_index()
        labels = [str(interval) for interval in value_counts.index]
    else:
        value_counts = df[col].value_counts().sort_index()
        labels = value_counts.index.astype(str)
 
    values = value_counts.values
    fig = go.Figure(data=[go.Pie(labels=labels, values=values, hole=0.4)])
    fig.update_layout(title=f"Donut Chart - {col}", width=500, height=400)
    fig.show()
```

```python
for i in range(len(df.columns)):
    for j in range(i + 1, len(df.columns)):
        plt.figure(figsize=(8, 6))
        sns.scatterplot(data=df, x=df.columns[i], y=df.columns[j], hue="Outcome", palette="coolwarm")
        plt.title(f"Scatter Plot - {df.columns[i]} vs {df.columns[j]}")
        plt.show()  
        plt.savefig(f"output_charts/scatter_{df.columns[i]}_vs_{df.columns[j]}.png")
        plt.close()
```

```python
plt.figure(figsize=(10, 5))
sns.heatmap(df.isnull(), cbar=False, cmap='plasma')
plt.title("Heatmap - Missing Values")
plt.show()  
plt.savefig("output_charts/heatmap_missing_values.png")
plt.close()
```
